```python
import numpy 
from matplotlib import pyplot
%matplotlib inline
# Create the time grid.
T = 100.0  # length of the time-interval
dt = 0.02  # time-step size
N = int(T / dt) + 1  # number of time steps
t = numpy.linspace(0.0, T, num=N)  # time grid
# Set the initial conditions.
z0 = 100.0  # altitude
b0 = 10.0  # upward velocity resulting from gust
zt = 100.0  # trim altitude
g = 9.81  # acceleration due to gravity

# Set the initial value of the numerical solution.
u = numpy.array([z0, b0])

# Create an array to store the elevation value at each time step.
z = numpy.zeros(N)
z[0] = z0
# Temporal integration using Euler's method.
for n in range(1, N):
    rhs = numpy.array([u[1], g * (1 - u[0] / zt)])
    u = u + dt * rhs
    z[n] = u[0]
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16

# Plot the solution of the elevation.
pyplot.figure(figsize=(9.0, 4.0))  # set the size of the figure
pyplot.title('Elevation of the phugoid over the time')  # set the title
pyplot.xlabel('Time [s]')  # set the x-axis label
pyplot.ylabel('Elevation [m]')  # set the y-axis label
pyplot.xlim(t[0], t[-1])  # set the x-axis limits
pyplot.ylim(40.0, 160.0)  # set the y-axis limits
pyplot.grid()  # set a background grid to improve readability
pyplot.plot(t, z, color='C0', linestyle='-', linewidth=2);
```

```python
import numpy
from matplotlib import pyplot

# the time-step sizes
dtValues = [0.1, 0.05, 0.01, 0.005, 0.001, 0.0001]
# the array to store error
errorArray = []

# initial conditions
# length of time interval
T = 100.0
# origin height
z0 = 100.0
# upward speed
b0 = 10.0
# trim height
zt = 100.0
# g
g = 9.81


# Euler method function
def euler_method(dt, z0, b0):
    # the number of points
    n = int(T/dt) + 1
    # set the initial value for the numerical solution
    u = numpy.array([z0, b0])
    # z to store the value
    z = numpy.zeros(n)
    z[0] = z0
    for i in range(n):
        z[i] = u[0] + dt * u[1]
        u[1] = u[1] + dt * (g - g * u[0] / zt)
        u[0] = z[i]
    return z


# function to calculate the analytical result
def analytical_method(dt, z0, b0):
    # the number of points
    n = int(T/dt) + 1
    t = numpy.arange(0.0, T + dt, dt)
    z_exact = (b0 * (zt / g) ** 0.5 * numpy.sin((g / zt) ** 0.5 * t) +
               (z0 - zt) * numpy.cos((g / zt) ** 0.5 * t) + zt)
    return z_exact


# function to calculate L1-norm error
def l1_error(dt, z, z_exact):
    error = dt*numpy.sum(numpy.abs(z-z_exact))
    return error


# calculate error
for dt in dtValues:
    # calculate numerical results and the analytical results
    z = euler_method(dt, z0, b0)
    z_exact = analytical_method(dt, z0, b0)
    # calculate the error
    error = l1_error(dt, z, z_exact)
    errorArray.append(error)

# plot the figure of error-dt
pyplot.figure(figsize=(6.0, 6.0))
pyplot.title('L1-norm error vs. time-step size')
pyplot.xlabel('time-step size')
pyplot.ylabel('L1-norm error')
pyplot.grid()
pyplot.loglog(dtValues, errorArray, color='C0', linestyle='--', marker='o')
pyplot.axis('equal')
pyplot.show()
```
